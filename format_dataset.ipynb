#### Load in clean dataset with normalised titles

```python
df = pd.read_csv("data/jobs_clean_jd.csv")
len(df)
```

```python
df.columns
```

```python
df = df[['title','location','remote_allowed','company_employee_count','Required_Skills','Educational_Requirements','Experience_Level','Preferred_Qualifications','company_industries','formatted_work_type','normalized_salary']].copy()
```

```python
import re
dataset = {'question':[], 'answer':[]}
for row in df.itertuples(index=False):
    title = row.title
    title = re.sub(r'\d', 'x', title)
    remote = "Yes" if row.remote_allowed == 1 else 'No'
    question = f"""<Job Title>
{title}

<Industry>
{row.company_industries}

<Location>
{row.location}

<Work Type>
{row.formatted_work_type}

<Remote Working Avaliability> 
{remote}

<Required Skills>
{row.Required_Skills}

<Educational Requirements>
{row.Educational_Requirements}

<Experience Level>
{row.Experience_Level}

<Preferred Qualifications>
{row.Preferred_Qualifications}

<Company Employee Count>
{row.company_employee_count}

The annual salary for this job is:
"""
    dataset['question'].append(question)
    dataset['answer'].append(row.normalized_salary)
```
