# Export Dataset

## Setup

```python
import pandas as pd
import kagglehub
```

## Read Data

```python
'''
Dataset from https://www.kaggle.com/datasets/arshkon/linkedin-job-postings/data.
'''
path = kagglehub.dataset_download("arshkon/linkedin-job-postings")
print("Path to dataset files:", path)
```

```python
df_jobs = pd.read_csv(path + '/postings.csv')
```

```python
df_jobs
```

```python
df_cos = pd.read_csv(path + '/companies/companies.csv')
```

```python
df_cos
```

```python
df_coinds = pd.read_csv(path + '/companies/company_industries.csv')
```

```python
df_coinds
```

```python
df_coemps = pd.read_csv(path + '/companies/employee_counts.csv')
```

```python
df_coemps
```

## Augment Jobs Data with Company Data

```python
'''
Join jobs data with companies data based on company ID.
Need to rename columns to avoid collision.
We ignore "company_size" because it's just a binned version of "employee_count" (see https://www.kaggle.com/datasets/arshkon/linkedin-job-postings/discussion/451340#2503021).
'''
df_jobs = df_jobs.merge(df_cos[['company_id', 'description', 'state', 'country', 'city', 'zip_code']].rename(
    columns={
        'description': 'company_description',
        'state': 'company_state',
        'country': 'company_country',
        'city': 'company_city',
        'zip_code': 'company_zip_code'
    }
), on='company_id', how='left')
```

```python
'''
Join jobs data with company industries data based on company ID.
Group by company_id first because each company can be associated with many industries. 
'''
df_jobs = df_jobs.merge(
    df_coinds.groupby('company_id').agg(', '.join).reset_index().rename(
        columns={'industry': 'company_industries'}
    ), 
    on='company_id',
    how='left'
)
```

```python
'''
Join jobs data with company employees data based on company ID.
Group by company_id first because the data has multiple rows per company. 
Pick only the most recent entry based on (largest) "time_recorded".
'''
df_jobs = df_jobs.merge(
    df_coemps[['company_id', 'employee_count']].groupby('company_id').first().reset_index().rename(
        columns={'employee_count': 'company_employee_count'}
    ), 
    on='company_id', 
    how='left'
)
```

## Remove Irrelevant Columns

```python
'''
To not waste time, we get rid of columns that are very likely irrelevant to our formulated problem.
'''
df_jobs = df_jobs.drop(
    labels=['job_posting_url', 'views', 'applies', 'expiry', 'closed_time', 
     'application_url', 'application_type', 'posting_domain', 'fips'],
    axis=1, errors='ignore'
)
```

## Export Raw Dataset

```python
df_jobs.to_csv('data/jobs.csv', index=False)
df_jobs
```
